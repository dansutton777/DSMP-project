# Clustering of the data

```python
# Read the CSV File
fake_transactions = pd.read_csv('fake_transactional_data.csv')

fake_transactions = fake_transactions.dropna()

# Add a new column 'Category' to the dataset
fake_transactions['Category'] = fake_transactions['to_randomly_generated_account'].apply(get_category)
fake_transactions.head()
```

```python
# Convert categorical variables to numerical using one-hot encoding

# Clustering for Home_Department category in this example
fake_transactions = fake_transactions.loc[fake_transactions['Category'] == 'Home_Department']
fake_transactions.head()
```

```python
# Convert to show date only
from datetime import datetime

# Load the fake_transactions dataset
df = fake_transactions

# Convert the "not_happened_yet_date" column to datetime format
df['not_happened_yet_date'] = pd.to_datetime(df['not_happened_yet_date'])

# Convert the "not_happened_yet_date" column to show date only
df['not_happened_yet_date'] = df['not_happened_yet_date'].dt.date

# Create TotalAmount colummn
df['TotalAmount'] = df['monopoly_money_amount']

# Create date variable that records recency
import datetime
snapshot_date = max(df['not_happened_yet_date']) + datetime.timedelta(days=1)

# Aggregate data by each "to_randomly_generated_account"
transactions = df.groupby(['from_totally_fake_account']).agg({
    'not_happened_yet_date': lambda x: (snapshot_date - x.max()).days,
    'Category': 'count',
    'TotalAmount': 'sum'})

# Rename columns
transactions.rename(columns = {'not_happened_yet_date': 'Recency',
                            'Category': 'Frequency',
                            'TotalAmount': 'MonetaryValue'}, inplace=True)
```

```python
transactions.head(100)
```

```python
# Manage skewness levels per each category
fig, ax = plt.subplots(1, 3, figsize=(15,3))
sns.distplot(transactions['Recency'], ax=ax[0])
sns.distplot(transactions['Frequency'], ax=ax[1])
sns.distplot(transactions['MonetaryValue'], ax=ax[2])
print("Skewness of variables (Home Department):")
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Manage skewneess by log & boxcox transformations
def manage_skewness(x):
    fig, ax = plt.subplots(2, 2, figsize=(5,5))
    sns.distplot(transactions[x], ax=ax[0,0])
    sns.distplot(np.log(transactions[x]), ax=ax[0,1])
    sns.distplot(np.sqrt(transactions[x]), ax=ax[1,0])
    sns.distplot(stats.boxcox(transactions[x])[0], ax=ax[1,1])
    plt.tight_layout()
    plt.show()
    
    print(transactions[x].skew().round(2))
    print(np.log(transactions[x]).skew().round(2))
    print(np.sqrt(transactions[x]).skew().round(2))
    print(pd.Series(stats.boxcox(transactions[x])[0]).skew().round(2))
```

```python
manage_skewness('Recency')
```

```python
# Apply cubic root transformation to the data
fig, ax = plt.subplots(1, 2, figsize=(10,3))
sns.distplot(transactions['MonetaryValue'], ax=ax[0])
sns.distplot(np.cbrt(transactions['MonetaryValue']), ax=ax[1])
plt.show()
print(transactions['MonetaryValue'].skew().round(2))
print(np.cbrt(transactions['MonetaryValue']).skew().round(2))
```

```python
# Apply the numbers and fix it  to a new dataframe
transactions_fix = pd.DataFrame()
transactions_fix["Recency"] = stats.boxcox(transactions['Recency'])[0]
transactions_fix["Frequency"] = stats.boxcox(transactions['Frequency'])[0]
transactions_fix["MonetaryValue"] = pd.Series(np.cbrt(transactions['MonetaryValue'])).values
transactions_fix.tail()
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize it such that all the means and std values are the same across the dataframe - then create new standardized dataframe
scaler = StandardScaler()
scaler.fit(transactions_fix)
transactions_normalized = scaler.transform(transactions_fix)
print(transactions_normalized.mean(axis = 0).round(2))
print(transactions_normalized.std(axis = 0).round(2))
```

```python
# Standardized dataframe of transaction RFM values
pd.DataFrame(transactions_normalized).head()
```

```python
from sklearn.cluster import KMeans

sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(transactions_normalized)
    # SSE to closest cluster centroid
    sse[k] = kmeans.inertia_ 

plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
# Bend of elbow method for this category of clustering appears to be at around 4 clusters
```

```python
# Fit the model with adjusted clusters based off of elbow method
model = KMeans(n_clusters=4, random_state=42)
model.fit(transactions_normalized)
model.labels_.shape
```

```python
# Assign cluster label for each account
transactions["Cluster"] = model.labels_
transactions.head()
```

```python
# Group cluster by mean RFM variables
transactions.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'MonetaryValue':['mean', 'count']}).round(1)
```

```python
df_normalized = pd.DataFrame(transactions_normalized, columns=['Recency', 'Frequency', 'MonetaryValue'])
df_normalized['ID'] = transactions.index
df_normalized['Cluster'] = model.labels_
df_normalized.head()
```

```python
# Melt The Data
df_nor_melt = pd.melt(df_normalized.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Recency','Frequency','MonetaryValue'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt.head()
```

```python
# Snake Plot
sns.lineplot('Attribute', 'Value', hue='Cluster', data=df_nor_melt).set_title('RFM Line Plot - Home Department')
```

```python
# Parallel Coordinates Plot
pd.plotting.parallel_coordinates(df_normalized.drop(columns=['ID']), 'Cluster', color=('red', 'blue', 'green')).set_title('RFM Parallel Coord Plot - Home Department')
plt.show()
```

```python
supermarket_cluster = fake_transactions.loc[fake_transactions['Category'] == 'Supermarket_Grocery']
supermarket_cluster.head()
```

```python
from datetime import datetime

# Load the fake_transactions dataset
df = supermarket_cluster

# Convert the "not_happened_yet_date" column to datetime format
df['not_happened_yet_date'] = pd.to_datetime(df['not_happened_yet_date'])

# Convert the "not_happened_yet_date" column to show date only
df['not_happened_yet_date'] = df['not_happened_yet_date'].dt.date

# Create TotalAmount colummn
df['TotalAmount'] = df['monopoly_money_amount']

# Create date variable that records recency
import datetime
snapshot_date = max(df['not_happened_yet_date']) + datetime.timedelta(days=1)

# Aggregate data by each "to_randomly_generated_account"
supermarket = df.groupby(['from_totally_fake_account']).agg({
    'not_happened_yet_date': lambda x: (snapshot_date - x.max()).days,
    'Category': 'count',
    'TotalAmount': 'sum'})

# Rename columns
supermarket.rename(columns = {'not_happened_yet_date': 'Recency',
                            'Category': 'Frequency',
                            'TotalAmount': 'MonetaryValue'}, inplace=True)
```

```python
supermarket.describe()
```

```python
from sklearn.cluster import KMeans

sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(supermarket)
    # SSE to closest cluster centroid
    sse[k] = kmeans.inertia_ 

plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(supermarket)
model.labels_.shape
```

```python
# Assign cluster label for each account
supermarket["Cluster"] = model.labels_
supermarket.head(100)
```

```python
supermarket.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'MonetaryValue':['mean', 'count']}).round(1)
```
